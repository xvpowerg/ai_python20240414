### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list3 = ["A","B","C","D","E"]
print(list3)
del list3[2]
print(list3)
del list3[1:3]
print(list3)
```

```python
list3 = ["A","B","C","D","E"]
del list3[-3:-1:1]
print(list3)
```

```python
list3 = ["A","B","C","D","E"]
del list3[-1:-3:-1]
print(list3)
```

## isinstance

isinstance(x,int) 代表判斷x是否為int型態

list若要進行排序得先挑選出相同型態資料。

請使用isinstance(資料,資料型態)方式於list中撈出特定資料。

list2a=[x for x in list2 if isinstance(x,int)]

這樣的寫法的目的

避免程式流程受到外面的干擾

流程單純化，list的資料變化就在list區域內進行

促進語法的可讀性提高!

```python
list1 = ['x',1,'y','z',5]
print(list1)
list2 = [ v for v in list1 if isinstance(v,int)]
print(list2)
list3 = [v for v in list1 if isinstance(v,str)]
print(list3)
```

## Dictionary

資料結構是由 key:value 所組成。

key 不能夠重複，否則會被後面的結果蓋過去。

可輸入 key 找尋您要找出來的值。

如果輸入的 key 不存在，那就會出現錯誤訊息。

key 只能是 字串 數字 tuple

```python
dic1 = {"a":100,"b":200,"c":300}
print(dic1["b"])
print(dic1["a"])
key = "x"
if key in dic1:
    print(dic1[key])
```

## key 與 value
+ key 不能於程式內改變：
  + 可以用數字、字串或者 tuple。
  + 不可以使用 list。
+ 如何找出所有的 key 與 value？
  + 您可以透過 dict1.keys( ) 這個方法找出所有的 key。
  + 您可以透過 dict1.values( ) 這個方法找出所有的 value。

```python
dic1 = {"a":100,"b":200,"c":300}

for k in dic1.keys():
    print(k,end = " ")
print()    
print("==============")
for k in dic1:
    print(k,end = " ")
print()
print("==============")
for v in dic1.values():
    print(v,end = " ")
print()
print("==============")    
```

## 新增與修改
+ 新增一筆資料
  + 請將 key 與 value 儲存至一個變數內。
  + 透過 dict.update(新增的資料) 這個方法的方式就可以新增。
+ 元素可否變更：
  + 以 = 指派方式指派給 key 就可以變更資料

```python
dict2 = {'ken':100,'vivin':95,'lucy':72}
print(dict2)
dict2["vivin"] = 77
print(dict2)
dict2["tom"] = 88#新增
print(dict2)
```

```python
dict2 = {'ken':100,'vivin':95,'lucy':72}
dict2.update({"lucy":88})
print(dict2)
dict2.update({"Ben":75,"vivin":12,"ken":65,"iris":77})
print(dict2)
```
