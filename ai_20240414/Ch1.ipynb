```python
"""
0~50 良好
51~100 普通
101~150 不健康
151~200  所有不健康
201~ 很不健康
"""
aqi = 201
if aqi >= 0 and aqi <= 50:
    print("良好")
else:
    if aqi >= 51 and aqi <= 100:
        print("普通")
    else:
        if aqi >= 101 and aqi <= 150:
            print("不健康")
        else:
            if aqi >= 151 and aqi <= 200:
                print("所有不健康")
            else:
                print("很不健康")
```
