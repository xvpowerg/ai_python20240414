# 作業：支出最大與最小

```python
priceList = []
priceSum = 0
for i in range(1,5):
    p = int(input(f"請輸入{i}個月的支出金額")) 
    priceSum += p
    priceList.append(p)
priceList.sort()
print(f"max{priceList[-1]}")
print(f"min{priceList[0]}")
print(f"sum{priceSum}")
print(priceList)

print(max(priceList))
print(min(priceList))
print(sum(priceList))
```

## Map

```python
list3 = [1,2,3,'',4,5,'']
mapObj = map(lambda x : x if x != '' else 0,list3)
print(list(mapObj))
```

```python
list3 = [1,2,3,'',4,5,'']
def sptoz(x):
    if x == '':
        return 0
    return x    
    
mapObj = map(sptoz,list3)
print(list(mapObj))
```

# Python 程式偵錯

+ Syntax Errors 語法錯誤
 + 該行語法錯誤，Python 無法編譯執行。
 + 修正語法錯誤
+ Exception 例外
 + 語法撰寫上沒有錯誤，執行時因資料邏輯或外部系統狀態發生錯誤， Python 會引發例外(Exception) 
 + 如果例外沒有被處理而傳遞到Python直譯器，程式中斷執行


```python
str1 = "AAA15"
try:
    print(int(str1))
except:
    print("請輸入數字")
print("錯誤後!")
```

## 使用raise拋例外

```python
def testScore(score):
    if score < 0 or  score> 100:
        #print("錯誤的成績")
        raise OverflowError("錯誤的成績")
    else:
        print("成績:",score)
    return score
    # 大於等於0 小於等於100 正確的成績
    # 不正確 就顯示錯誤的成績
try:
    s = testScore(-10)    
    print("計算成績:",s)
except OverflowError as ex:
    print(ex)

```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
str1 = "Ken,Vivin,Lucy"
nameList =  str1.split(",")
print(nameList)
print(nameList[1])
```

```python
import re
text = 'Ken, Vivin...Lucy'
result = re.split(r"[,. ]+",text)
print(result)
result2 = re.split(r"[,. ]+",text,1)
print(result2)
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。

```python
import re
inputStr = "hello 123 world 456 8"
newStr = inputStr.replace("123","789")
print(newStr)
newStr2 = re.sub("\d+","258",inputStr)
print(newStr2)
newStr3 = re.sub("\d+","258",inputStr,count = 1)
print(newStr3)
```

```python
import re
str1 = "This is Python.That is Java,This is SQLite"
newString = str1.replace("is","IS")
print(newString)
newStr2 = re.sub(r" is "," IS ",str1)
print(newStr2)
newStr3 = re.sub(r"\bis\b","IS",str1)
print(newStr3)
```

```python
import re
pattern = re.compile(r"\d+")#固定樣式
str1 = "hello 123 world 456 8"
result1 = pattern.findall(str1)
print(result1)
result2 = pattern.split(str1)
print(result2)
result3 =  pattern.sub("@",str1)
print(result3)
```
